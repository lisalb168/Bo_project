## 1. Stratified Sampling

As there are only 8% of default applications, the dataset is highly imbalanced. Use stratified sampling in sk-learn to do train test split helps maintaining the default proportions within the training and test groups. 

```python
from sklearn.model_selection import StratifiedShuffleSplit
```

```python
X = data_all.iloc[:,1:]
y = data_all['TARGET']

sss = StratifiedShuffleSplit(n_splits=2, test_size=0.2, random_state=123)
sss.get_n_splits(X, y)

for train_index, test_index in sss.split(X, y):
    print("TRAIN:", train_index, "TEST:", test_index)
    
X_train_strat, X_test_strat = X.loc[train_index, :], X.loc[test_index, ]
y_train_strat, y_test_strat = y[train_index], y[test_index]
print("Default/Non Default counts in Training data \n", y_train_strat.value_counts())
print("Default/Non Default percent in Training data \n", y_train_strat.value_counts(normalize=True))
print("Default/Non Default counts in Test data \n", y_test_strat.value_counts())
print("Default/Non Default percent in Test data \n", y_test_strat.value_counts(normalize=True))

len(data_all) == len(X_train_strat) + len(X_test_strat)
```

## 2. Random Downsampling

The model performance on imbalanced dataset is poor, use downsampling to balance the majority and minority groups into approximately equal sizes before modeling.

```python
from sklearn.model_selection import train_test_split
```

```python
# train test split using test_size = 0.2 
X_train, X_test, y_train, y_test= train_test_split(X, y, test_size=0.2, random_state=123)

X_train_0 = X_train.loc[y_train == 0,:].copy().sample(frac = 0.09)
X_train_1 = X_train.loc[y_train == 1,:].copy()
print("Shape of X train in the downsampled majority class \n", X_train_0.shape)

X_train_0['target'] = 0
X_train_1['target'] = 1
print("Downsampled y train counts \n", X_train_0.target.value_counts())

# use frac = 1 to randomize the rows
X_train_new = X_train_0.append(X_train_1).sample(frac = 1)
print("Balanced y train counts \n", X_train_new.target.value_counts())
print("Balanced y train percent \n", X_train_new.target.value_counts(normalize=True))

X_train_new_x = X_train_new.drop(columns='target')
y_train_new = X_train_new['target']

print("Length of balanced X train \n", len(X_train_new_x))
print("Length of balanced y train \n", len(y_train_new))

print("y test counts (still imbalanced) \n", y_test.value_counts())
print("y test percent (still imbalanced) \n", y_test.value_counts(normalize=True))
```

## Define functions to run each step in the pipeline

```python
def CrossValidation(searchmethod, params, Kfold, sampling, 
                    sampled_train_X, sampled_train_y, classifier, niter=5, verbose=1, random_state=123):
    
    best = [] # used to save the best parameter combinations
    
    if searchmethod == 'grid' and classifier == 'RandomForest':
        clf = RandomForestClassifier(random_state=random_state)
        cv_clf = GridSearchCV(clf, params, cv=Kfold, verbose=verbose, n_jobs=-1) 
        cv_clf.fit(sampled_train_X, sampled_train_y)
        best_n = cv_clf.best_params_['n_estimators']
        best_d = cv_clf.best_params_['max_depth']
        print(cv_clf.best_estimator_)
        print(cv_clf.best_params_)
        print(cv_clf.best_score_)
        print(cv_clf.cv_results_)
        print(cv_clf.best_index_)
        print("Best number of trees and depth are: ", best_n, best_d)
        best = [best_n, best_d]
        return best

    elif searchmethod == 'rand' and classifier == 'RandomForest':
        clf = RandomForestClassifier(random_state=random_state)
        cv_clf = RandomizedSearchCV(clf, params, cv=Kfold, n_iter=niter, verbose=verbose, n_jobs=-1) 
        cv_clf.fit(sampled_train_X, sampled_train_y)
        best_n = cv_clf.best_params_['n_estimators']
        best_d = cv_clf.best_params_['max_depth']
        print(cv_clf.best_estimator_)
        print(cv_clf.best_params_)
        print(cv_clf.best_score_)
        print(cv_clf.cv_results_)
        print(cv_clf.best_index_)
        print("Best number of trees and depth are: ", best_n, best_d)
        best = [best_n, best_d]
        return best
    
    elif searchmethod == 'rand' and classifier == 'Logistic':
        clf = LogisticRegression(random_state=random_state)
        cv_clf = RandomizedSearchCV(clf, params, cv=Kfold, n_iter=niter, verbose=verbose, n_jobs=-1) 
        cv_clf.fit(sampled_train_X, sampled_train_y)
        best_c = cv_clf.best_params_['C']
        best_penalty = cv_clf.best_params_['penalty']
        print(cv_clf.best_estimator_)
        print(cv_clf.best_params_)
        print(cv_clf.best_score_)
        print("Best C and penalty function are: ", best_c, best_penalty)
        best = [best_c, best_penalty]
        return best
        
    elif searchmethod == 'rand' and classifier == 'KNN':
        clf = KNeighborsClassifier()
        cv_clf = RandomizedSearchCV(clf, params, cv=Kfold, n_iter=niter, verbose=verbose, n_jobs=-1) 
        cv_clf.fit(sampled_train_X, sampled_train_y)
        best_k = cv_clf.best_params_['n_neighbors']
        print(cv_clf.best_estimator_)
        print(cv_clf.best_params_)
        print(cv_clf.best_score_)
        print("Best number of neighbors is: ", best_k)
        best = [best_k]
        return best
        
    elif searchmethod == 'rand' and classifier == 'LinearSVM':    
        clf = svm.LinearSVC(random_state=random_state, tol=1e-5)
        cv_clf = RandomizedSearchCV(clf, params, cv=Kfold, n_iter=niter, verbose=verbose, n_jobs=-1) 
        cv_clf.fit(sampled_train_X.values, sampled_train_y.values)
        best_c = cv_clf.best_params_['C']
        best_penalty = cv_clf.best_params_['penalty']
        print(cv_clf.best_estimator_)
        print(cv_clf.best_params_)
        print(cv_clf.best_score_)
        print("Best C and penalty are: ", best_c, best_penalty) 
        best = [best_c, best_penalty]
        return best       

    elif searchmethod == 'rand' and classifier == 'KernelSVM':    
        clf = svm.SVC(random_state=random_state)
        cv_clf = RandomizedSearchCV(clf, params, cv=Kfold, n_iter=niter, verbose=verbose, n_jobs=-1) 
        cv_clf.fit(sampled_train_X.values, sampled_train_y.values)
        best_c = cv_clf.best_params_['C']
        best_kernel = cv_clf.best_params_['kernel']  
        best_gamma = cv_clf.best_params_['gamma']
        print(cv_clf.best_estimator_)
        print(cv_clf.best_params_)
        print(cv_clf.best_score_)
        print("Best C, kernel and gamma are: ", best_c, best_kernel, best_gamma) 
        best = [best_c, best_kernel, best_gamma]
        return best       
    
    elif searchmethod == 'rand' and classifier == 'Xgboost': 
        clf = XGBClassifier(objective='binary:logistic', verbosity=verbose, booster='gbtree', tree_method='auto', 
                            subsample=1, colsample_bytree=1, colsample_bylevel=1, colsample_bynode=1, 
                            reg_alpha=0, reg_lambda=1, scale_pos_weight=1, base_score=0.5, random_state=random_state)
        cv_clf = RandomizedSearchCV(clf, params, cv=Kfold, n_iter=niter, verbose=verbose, n_jobs=-1) 
        cv_clf.fit(sampled_train_X, sampled_train_y)
        best_n = cv_clf.best_params_['n_estimators']
        best_lr = cv_clf.best_params_['learning_rate'] 
        best_d = cv_clf.best_params_['max_depth']
        best_gamma = cv_clf.best_params_['gamma']
        print(cv_clf.best_estimator_)
        print(cv_clf.best_params_)
        print(cv_clf.best_score_)
        print("Best number of trees, learning rate, depth, gamma are: ", best_n, best_lr, best_d, best_gamma)
        best = [best_n, best_lr, best_d, best_gamma]
        return best    
    
    elif searchmethod == 'rand' and classifier == 'LightGB': 
        clf = LGBMClassifier(colsample_bytree=1, subsample=1, reg_alpha=0, reg_lambda=1, verbose=verbose)
        cv_clf = RandomizedSearchCV(clf, params, cv=Kfold, n_iter=niter, verbose=verbose, n_jobs=-1) 
        cv_clf.fit(sampled_train_X, sampled_train_y)
        best_n = cv_clf.best_params_['n_estimators']
        best_lr = cv_clf.best_params_['learning_rate'] 
        best_d = cv_clf.best_params_['max_depth']
        print(cv_clf.best_estimator_)
        print(cv_clf.best_params_)
        print(cv_clf.best_score_)
        print("Best number of trees, learning rate, depth are: ", best_n, best_lr, best_d)
        best = [best_n, best_lr, best_d]
        return best       
    
    else:
        print("No search method and classifier combination found!")
        return best
```

```python
# use the best parameters to fit a model on the training data and predict on the test data

def best_param_fit_model(classifier, sampled_train_X, sampled_train_y, sampled_x_test, sampled_y_test, 
                         best_n, best_d, best_c, best_penalty, 
                         best_k, best_kernel, best_gamma, best_lr, random_state=123):
    if classifier == 'RandomForest':
        clf = RandomForestClassifier(n_estimators = best_n, max_depth = best_d, random_state = random_state) 
    elif classifier == 'KNN':
        clf = KNeighborsClassifier(n_neighbors = best_k) 
    elif classifier == 'Logistic':
        clf = LogisticRegression(C = best_c, penalty = best_penalty, random_state = random_state) 
    elif classifier == 'LinearSVM':
        clf = svm.LinearSVC(C = best_c, penalty = best_penalty, random_state = random_state)
    elif classifier == 'KernelSVM':
        clf = svm.SVC(C = best_c, kernel = best_kernel, gamma = best_gamma, random_state = random_state)
    elif classifier == 'Xgboost':
        clf = XGBClassifier(n_estimators = best_n, learning_rate = best_lr, max_depth = best_d, 
                            gamma = best_gamma, objective='binary:logistic', booster='gbtree', tree_method='auto', 
                            subsample=1, colsample_bytree=1, colsample_bylevel=1, colsample_bynode=1, 
                            reg_alpha=0, reg_lambda=1, scale_pos_weight=1, base_score=0.5, 
                            random_state = random_state)  
    elif classifier == 'LightGB':
        clf = LGBMClassifier(n_estimators = best_n, learning_rate = best_lr, max_depth = best_d, 
                            colsample_bytree=1, subsample=1, reg_alpha=0, reg_lambda=1)  
    else:
        print("No search method and classifier combination found!")
        
    model = clf.fit(sampled_train_X, sampled_train_y)
    
    preds = clf.predict(sampled_x_test)
    
    features = sampled_train_X.columns
    
    if classifier not in ['LinearSVM','KernelSVM']:
        probs = clf.predict_proba(sampled_x_test)
    else:
        probs = clf.decision_function(sampled_x_test)
    
    if classifier in ['RandomForest','Xgboost','LightGB']:
        importances = clf.feature_importances_
        indices = np.argsort(importances)
    else:
        importances = 0
        indices = 0
    
    confusion_matrix = sklearn.metrics.confusion_matrix(sampled_y_test, preds, labels=None, sample_weight=None)
    report = classification_report(sampled_y_test, preds)
    
    if classifier not in ['LinearSVM','KernelSVM']:
        auc = roc_auc_score(sampled_y_test, probs[:,1])
    else:
        auc = roc_auc_score(sampled_y_test, probs)
    
    accuracy = float(np.sum(preds==sampled_y_test))/sampled_y_test.shape[0] 
    print("accuracy: %f" % (accuracy))

    results = []
    results = [accuracy, preds, probs, features, importances, indices, model, confusion_matrix, report, auc]
    return results
```

```python
# plot the top M feature importances
def imp_plot(importances, indices, features, M=10):
    plt.title('Feature Importances')
    plt.barh(range(M), importances[indices][-M:], color='b', align='center')
    plt.yticks(range(M), [features[i] for i in indices[-M:]])
    plt.xlabel('Relative Importance')
    plt.show()
```

```python
def roc_p_r(model, sampled_x_test, sampled_y_test, preds, probs, classifier):
    average_precision = average_precision_score(sampled_y_test, preds)
    print('Average precision-recall score: {0:0.2f}'.format(average_precision))
    #disp = plot_precision_recall_curve(model, sampled_x_test, sampled_y_test)
    #disp.ax_.set_title('2-class Precision-Recall curve: '
    #                   'AP={0:0.2f}'.format(average_precision))
    #plt.show()

    # Compute ROC curve and ROC area for each class
    if classifier in ['LinearSVM','KernelSVM']:
        y_score = model.decision_function(sampled_x_test)
    else:
        y_score = probs[:,1]
        
    fpr, tpr, thresholds = roc_curve(sampled_y_test, y_score)
    print("FPR: ", fpr)
    print("TPR: ", tpr)
    print("Threshold: ", thresholds)
    roc_auc = auc(fpr, tpr)
    print("AUC: ", roc_auc)
    
    plt.figure()
    lw = 2
    plt.plot(fpr, tpr, color='darkorange', lw=lw, label='ROC curve (area = %0.2f)' % roc_auc)
    plt.plot([0, 1], [0, 1], color='navy', lw=lw, linestyle='--')
    plt.xlim([0.0, 1.0])
    plt.ylim([0.0, 1.05])
    plt.xlabel('False Positive Rate')
    plt.ylabel('True Positive Rate')
    plt.title('Receiver Operating Characteristic')
    plt.legend(loc="lower right")
    plt.show()
```

### 1.1.1b Hyperparameter tuning using RandomizedSearchCV

RandomizedSearchCV approach will randomly pick a fixed pre-specified number of combinations of parameters in the specified spaces and compute the K-fold CV accuracy scores and pick the best parameter combination coresponding to the highest average accuracy. This approach gives the modeler the flexibility to choose how many parameter combinations to search on, and thus can be less time consuming, while at the same time picking the best parameter combinations within a large range of pre-specified parameter spaces.
